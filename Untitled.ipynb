```julia
streams_1 = joinpath(ENV["CFS"], "nstaff", "blaschke", "slurm_3", "cori_2.stream")
```

```julia
x = open(ZlibDecompressorStream, streams_1, "r") do stream
    contents = String[]
    while !eof(stream)
        size = read(stream, Int64)
        entry = ""
        for i=1:size
            entry *= read(stream, Char)
        end
        push!(contents, entry)
    end
    contents
end

dt = @chain x[end] begin
    JSON.parse(_; dicttype=Dict{Symbol, Any})
    _[:submitTime]
    unix2datetime
end
```

```julia
ids = []
for i in x
    try
        push!(ids, JSON.parse(i; dicttype=Dict{Symbol, Any})[:jobId])
    catch
    end
end
```
