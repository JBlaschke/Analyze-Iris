```julia
ue_seconds(x::String)  = Int64(floor(datetime2unix(DateTime(x, "yyyy-mm-ddTHH:MM:SS")))) #.instant.periods.value
start(r::Reservation)  = ue_seconds(r.starttime)
finish(r::Reservation) = ue_seconds(r.endtime)
```

```julia
moving_average(vs,n) = [sum(@view vs[i:(i+n-1)])/n for i in 1:(length(vs)-(n-1))]
```

```julia
function sacct_get_jobs(res::Reservation)
    sacct_cmd = `sacct --nodelist=$(res.nodelist) --starttime=$(res.starttime) --endtime=$(res.endtime) -nXPo AdminComment`
    lines = open(sacct_cmd) do f
        readlines(f)
    end
    return map(x->JSON.parse(x), lines)
end
```

```julia
lx31_jobs = vcat(sacct_get_jobs(lx31_1), sacct_get_jobs(lx31_1a), sacct_get_jobs(lx31_2))
lv07_jobs = vcat(sacct_get_jobs(lv07_1), sacct_get_jobs(lv07_2), sacct_get_jobs(lv07_3), sacct_get_jobs(lv07_4))
p203_jobs = vcat(sacct_get_jobs(p203_1), sacct_get_jobs(p203_2), sacct_get_jobs(p203_3))

length(lx31_jobs) + length(lv07_jobs) + length(p203_jobs)
```

```julia
struct Job
    startTime::Int64
    endTime::Int64
    submitTime::Int64
    jobAccount::String
    allocCpus::Int64
    qos::String
    
    function Job(job::Dict{String, Any})
        new(
            job["startTime"],
            job["endTime"],
            job["submitTime"],
            job["jobAccount"],
            job["allocCpus"],
            job["qos"]
        )
    end
end

start(j::Job)    = j.startTime
finish(j::Job)   = j.endTime
duration(j::Job) = finish(j) - start(j)
cpus(j::Job)     = j.allocCpus
submit(j::Job)   = j.submitTime
waittime(j::Job) = start(j) - submit(j)
```

```julia
jobs_lcls   = filter(j -> j.jobAccount=="lcls", Job.(jobs))
jobs_nstaff = filter(j -> j.jobAccount=="nstaff", Job.(jobs))
jobs_other  = filter(j -> j.jobAccount!="nstaff" && j.jobAccount!="lcls", Job.(jobs))
```

```julia
compute = zeros(Int64, finish_all - start_all + 100000)
waiting = zeros(Int64, finish_all - start_all + 100000)

for job in jobs_lcls
    start_job = start(job)  - start_all + 1
    end_job   = finish(job) - start_all

    n_cpus = cpus(job)
    for sec = start_job:end_job
        compute[sec] += n_cpus
    end
    
    submit_job = submit(job) - start_all + 1
    
    n_wait = cpus(job)
    for sec = submit_job:start_job
        waiting[sec] += n_wait
    end
end
```

```julia
compute_preemt = zeros(Int64, finish_all - start_all + 100000)

for job in jobs_nstaff
    start_job = start(job)  - start_all + 1
    end_job   = finish(job) - start_all

    n_cpus    = cpus(job)
    for sec = start_job:end_job
        compute_preemt[sec] += n_cpus
    end
end
```
